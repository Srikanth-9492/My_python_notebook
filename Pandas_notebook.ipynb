**loc()**

```python
#loc is used to locate rows
#df.loc[start:end:step]
#print the entire 1sr row of my dataframe
df.loc[0]
#print the values of index rows from 100 to 105
#df.loc[start:end]
df.loc[100:105]
#read the rows with even and odd indexes up to the 100th row
df.loc[0:100:2] #even index rows
df.loc[1:100:2] #odd index rows
```

```python
#display the even index position characters of a particular column ---> 'total_bedroom'
df.loc[0:100:2,'total_bedrooms']
df['total_bedrooms'].loc[0:20:2]
```

```python
#display even index position values of a specific columns--->'total_rooms','total_bedrooms', 'population'
df[['total_rooms','total_bedrooms', 'population']].loc[0:100:2]
df.loc[0:20:2,['total_rooms','total_bedrooms', 'population']]
```

```python
#Applying the condition for the column
#print the even values of the column total_bedrooms
df['total_bedrooms'] % 2 == 0
```

```python
df.loc[df['total_bedrooms']%2==0]['total_bedrooms']
```

```python
df_even = df[df['total_bedrooms']%2==0]['total_bedrooms']
df_even.head()
```

```python
#locate the rows in between
df.loc[[6500,2500,7000]]
```

**Iloc**---> passes the axis number of rows and columns

```python
df.columns
```

```python
#syntax--->df.iloc[row_index,column_index]
#print the first 3 rows of following columns 'latitude', 'housing_median_age', 'total_rooms','total_bedrooms'
df.iloc[0:3,1:5]
df.iloc[1,3]
df.iloc[3,6]
```

```python
df.iloc[0:100:2,1:5]
```

```python
#what is difference between loc and iloc
#loc supports the filtering where as iloc doesnot support filtering
#we can apply condition in loc and we cannot apply condition in iloc.
df.iloc[df['total_bedrooms']%2==0]['total_bedrooms']
```

```python
#locate the rows of  specific columns, 'population', 'households', 'housing_median_age' [for ex : locate first 10 rows]
df.loc[0:10,['population', 'households', 'housing_median_age']]
```
