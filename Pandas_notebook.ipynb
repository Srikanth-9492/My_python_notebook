```python
#How many data structures do we have in pandas?
#There are 3 types
#1.SERIES:- series are nothing but 1D labelled array's
#How can we build a series?
```

```python
import numpy as np
a = np.array([1,2,3,4])
print(a)
```

```python
sr = pd.Series(a)
print(sr)
```

```python
#How can we change the labelles?
sr = pd.Series(a,index=['a','b','c','d'])
print(sr)
```

```python
#create a series from the dictionary?
d = {
    "name":"Sai",
    "Course": "DS",
    "Duration": "4months"
}
sr = pd.Series(d)
print(sr)
#How can we represent a series in pandas?
print(type(sr))
#ans) <class 'pandas.core.series.Series'>
```
