```python
#What is heat map?
#it is data visualization technique that shows the magnitude phenomenon
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = np.random.randn(5,5)
#create the heat map
sns.heatmap(data, annot=True,cmap='YlGnBu')
plt.title("Basic heatmap with random numbers")
plt.show()
```

```python
df = pd.DataFrame({
    'maths':[88,92,79,93],
    'Science':[84,94,75,90],
    'English':[90,85,80,88]
},index=['Andrew','charlie','David','sam'])
plt.figure(figsize=(5,3))
sns.heatmap(df,annot=True,cmap='magma')
plt.title("student scores heatmap")
plt.show()
```

```python
df2 = sns.load_dataset('iris')
#compute the co-relation values
corr =df2.corr(numeric_only=True)
corr
plt.figure(figsize=(5,3))
sns.heatmap(corr,annot=True,cmap='Blues',center=0)
plt.title('heatmap of corr')
plt.show()
```

```python
#create a heatmap showing average tip amount by day and time by using tips data
df1 = sns.load_dataset('tips')
pivot = df1.pivot_table(values='tip',index='day',columns='time',aggfunc='mean')
sns.heatmap(pivot,annot=True,cmap='inferno',center=0)
plt.title("Average tip by day and time")
plt.show()
```

Distributional Representation

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
#displot:- supports to visualize the distribution of your datasets
#plot graph between total_bill sccording to time wise
```

```python
sns.displot(data=df, x='total_bill',col='time',kde=True)  #kde:-kernal distribution estimation.
plt.show()
```

```python
#Emperical cummulative distribution function:-- ECDF
sns.displot(data=df, x='total_bill',col='time',kind='ecdf',hue='smoker')
```

**Distribution of a categorical value**

```python
sns.scatterplot(data=df, x='total_bill',y='tip',hue='smoker')
```

```python
#to see a clear distribution of a categorical data----> catplot
sns.catplot(data=df,kind='swarm',x='day',y='total_bill',hue='smoker')
#sns.catplot(data=df,kind='box',x='day',y='total_bill',hue='smoker')
```

```python
sns.catplot(data=df,kind='swarm',x='sex',y='total_bill',hue='smoker')
```

**Statistical Estimation plot**

```python
#how can we enhance a scatterplot by including a linear regression model
# 'lmplot'
sns.lmplot(data=df,x='total_bill',y='tip',col='time',hue='smoker')
```
