**Data Visualization**

```python
#what is the purpose of data visualization?
#to better understand of the data we plot different graphs or figure
#In python there are 2 packages used to plot graphs
#1.matplotlib
#2.seaborn
```

Anatomy of a plot

```python
#what are all different parts do we have in a plot?
#1. co-ordinates (x-axis, y-axis)
#2. labels (x label, y label)
#3. title
#4. Ticks
#5. Grids
#6. legends
#7. figure
```

```python
import matplotlib.pyplot as plt
```

```python
#plot a figure on the heights of a classroom
heights = [160,140,150,180,160,170]
plt.plot(heights)
plt.show()
#Note:- In matplotlib the default is a line plot
```

```python
heights = [160,140,150,180,160,170]
weights = [65,70,68,64,72,90]
#plot a line plot between height and weight
plt.plot(heights, weights)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
import numpy as np
x = np.array([1,9])
y = np.array([3,10])
plt.plot(x,y)
plt.show()
```

```python
#instead of the line plot draw the scatter plot
plt.plot(x,y,'o')   #[o,x,*,p,P,+,s,D,d,h,H,v]
plt.show()
```

```python
a = np.array([1,2,3,4,5,6])
b = np.array([3,8,1,5,10,7])
#draw a plot based on the above array by using different symbols
plt.plot(a,b,'P')
plt.show()
```

Apply colours to the markers

```python
a = np.array([1,2,3,4,5,6])
b = np.array([3,8,1,5,10,7])
plt.plot(a,b,'d:y')
plt.show()
```

Applying the different formatting lines

```python
#1.solid line --> -
#2.Dotted line--> :
#3.Dashed line--> --
#4.dash dot line--> -.
```

```python
a = np.array([1,2,3,4,5,6])
b = np.array([3,8,1,5,10,7])
plt.plot(a,b,'d-y')   #solid line
plt.show()
```

```python
a = np.array([1,2,3,4,5,6])
b = np.array([3,8,1,5,10,7])
plt.plot(a,b,'d-.y') #dash dot line
plt.show()
```

```python
a = np.array([1,2,3,4,5,6])
b = np.array([3,8,1,5,10,7])
plt.plot(a,b,'d:y')  #dotted line
plt.show()
```

```python
a = np.array([1,2,3,4,5,6])
b = np.array([3,8,1,5,10,7])
plt.plot(a,b,'d--r')   #Dashed line
plt.show()
```

```python
#increase the size of the marker from the above plot ---> use 'ms'
plt.plot(a,b,'o-r',ms=10)
plt.show()
```

```python
#give the different marker edge color for above plot
plt.plot(a,b,'p-.r',ms=10,mec='b')
plt.show()
```

```python
#fill the marker color with different color
plt.plot(a,b,'p-.r',ms=10,mec='b',mfc='#c71875')
plt.show()
```

```python
#plot a graph between exam entry number and the marks'
x = [1,2,3,4,5,6,7,8,9]
y = [2,1,4,6,6,5,8,9,10]
plt.plot(x,y,'P')
plt.xlabel('entry number')
plt.ylabel('marks in exam')
plt.title('exam entry number vs marks',)
#use the grids()--> locate exact location of data point
#plt.grid()
#specify the grids according to x-axis and y-axis
#plt.grid(axis='y')
#plt.grid(axis='x')
#change the linestyle of the grid
plt.grid(axis='y',linestyle='--',linewidth='1',color='#e05d0b')
plt.show()
```

**Scatter Plot**

```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,1,4,6,6,5,8,9,10]
#draw a scatter for the above data
plt.scatter(x,y,color='#8fd411')
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,1,4,6,6,5,8,9,10]
#draw a scatter plot with different size and color
plt.scatter(x,y,color='#8fd411',s=100)
plt.show()
```

```python
x_female = [4.6,5.2,5.4,4.9,5.1,5.8,4.8]
y_female = [63,40,35,58,54,56,58]
plt.scatter(x_female,y_female,color='#940f70',s=60)
plt.show()
```

```python
#change the size of the scatter according to the weight of female
x_female = [4.6,5.2,5.4,4.9,5.1,5.8,4.8]
y_female = [63,40,35,58,54,56,58]
plt.scatter(x_female,y_female,color='#940f70',s=y_female)
plt.show()
```
