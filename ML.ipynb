```python
data.plot(x='Hours',y='Scores',style='o')
plt.title('Hours vs Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
x = data.loc[:,['Hours']].values
y = data.loc[:,['Scores']].values
```

```python
y
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=45)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
#fit the training data in model
regressor.fit(X_train,y_train)  #--->entire magic happens here, where loss function and gradienr descent computed here
```

```python
#m and c
#c is known as interception
print(regressor.intercept_) #c
```

```python
#m is nothing but slope --->co-effiecient
print(regressor.coef_)
```

```python
#y (predicted score) = 9.72*(hours) + 1.43
regressor.predict([[7.5]]) #perils of extra polations
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df
```

```python
#R2_score
from sklearn.metrics import r2_score
print('R2_score: ',r2_score(y_test,y_pred))
```
